# Localization Overview

Localization is the process of figuring out **where a robot is** within a world or environment.  
Because of **noisy sensors** and **uncertain motion**, a robot cannot know its position with absolute certainty.  

Instead of saying *‚Äúthe robot is exactly at x = 3‚Äù*, we represent its position as a **belief distribution** ‚Äî a probability distribution over all possible locations.  
Example:  
- 10% probability at x = 1  
- 20% probability at x = 3  
- ‚Ä¶ and so on.  

This belief distribution captures how confident the robot is about each possible location.

---

## Steps in Localization

1. **Measurement Update (Sense Step)**  
   - The robot takes a sensor measurement (e.g., detects a door, wall, or landmark).  
   - If the sensor reading *matches* a location ‚Üí probability increases.  
   - If it doesn‚Äôt match ‚Üí probability decreases.  
   üëâ This is what we coded with `pHit` and `pMiss`.

2. **Motion Update (Move Step)**  
   - The robot moves forward or backward.  
   - The belief distribution shifts in the direction of motion.  
   - Because movement is uncertain, the probabilities ‚Äúspread out.‚Äù  
   üëâ This is modeled using a convolution of probabilities.

# Markov Localization (Bayes Filter) ‚Äî AI for Robotics Practice

This notebook implements **Markov Localization**, a special case of Bayes Filters, 
to estimate a robot's position in a 1D world.  

- **World**: 5 cells (`red`, `green`, `green`, `red`, `red`)  
- **Goal**: Update the belief distribution as the robot senses and moves.  
- **Methods**:  
  - Measurement update (`sense`)  
  - Motion update (`move`)  
- **Visualization**: Belief distribution bar plots at each step.  

This work is part of my preparation for **graduate studies in Robotics and AI**.

```python
# World (1D corridor)
world = ['red', 'green', 'green', 'red', 'red']

# Initial belief: uniform distribution
p = [1.0 / len(world)] * len(world)

# Sensor accuracy
pHit = 0.6
pMiss = 0.2

# Motion accuracy
pExact = 0.8
pOvershoot = 0.1
pUndershoot = 0.1

print("Initial Belief:", p)
```

```python
def sense(p, Z):
    """Measurement update (Bayes rule)"""
    q = []
    for i in range(len(p)):
        hit = (Z == world[i])
        q.append(p[i] * (hit * pHit + (1 - hit) * pMiss))
    # Normalize
    s = sum(q)
    return [i / s for i in q]

def move(p, U):
    """Motion update (Markov assumption)"""
    q = [0.0] * len(p)
    for i in range(len(p)):
        q[(i+U) % len(p)]     += p[i] * pExact
        q[(i+U-1) % len(p)]   += p[i] * pOvershoot
        q[(i+U+1) % len(p)]   += p[i] * pUndershoot
    return q
```

```python
measurements = ['green', 'green', 'green', 'green', 'green']
motions = [1,1,1,1,1]

beliefs = [p.copy()]  # store all distributions

for k in range(len(measurements)):
    p = sense(p, measurements[k])
    p = move(p, motions[k])
    beliefs.append(p.copy())

print("Final Belief Distribution:", beliefs[-1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
for i, b in enumerate(beliefs):
    plt.figure()
    plt.bar(range(len(b)), b, tick_label=world)
    plt.title(f"Step {i}: Belief Distribution")
    plt.ylim(0,1)
    plt.xlabel("World Cells")
    plt.ylabel("Probability")
    plt.show()
```
